```python
import os
import glob
import random
import shutil
import pickle

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ultralytics import YOLO
from PIL import Image
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
dataset_path = "/content/drive/MyDrive/Rice.v2i.yolov8"
data_yaml = os.path.join(dataset_path,"data.yaml")

print("Dataset path:",dataset_path)
```

```python
train_imgs = glob.glob(dataset_path+"/train/images/*.jpg")

samples = random.sample(train_imgs,5)

plt.figure(figsize=(14,4))

for i,img in enumerate(samples):

    im = Image.open(img)

    plt.subplot(1,5,i+1)
    plt.imshow(im)
    plt.axis("off")

plt.suptitle("Sample Rice Images from Dataset")

plt.show()
```

```python
models = {

"YOLOv8":"yolov8n.pt",

"YOLOv10":"yolov10n.pt",

"YOLOv11":"yolo11n.pt"

}
```

```python
results_summary = []

for name,weight in models.items():

    print("Training:",name)

    model = YOLO(weight)

    model.train(

        data=data_yaml,

        epochs=80,

        patience=10,

        imgsz=640,

        batch=16,

        device=0,

        workers=4,

        name=name

    )

    metrics = model.val(data=data_yaml)

    results_summary.append({

        "Model":name,

        "mAP50":metrics.box.map50,

        "Precision":metrics.box.mp,

        "Recall":metrics.box.mr

    })
```

```python
df_compare = pd.DataFrame(results_summary)
print(df_compare)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x="Model",y="mAP50",data=df_compare)

plt.title("YOLO Model Comparison (mAP50)")

plt.show()
```

```python
df = pd.read_csv("runs/detect/YOLOv8/results.csv")

plt.figure(figsize=(10,5))

plt.plot(df['train/box_loss'],label='Train Loss')

plt.plot(df['val/box_loss'],label='Validation Loss')

plt.legend()

plt.title("Training vs Validation Loss")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df['metrics/precision(B)'],label="Precision")
plt.plot(df['metrics/recall(B)'],label="Recall")

plt.legend()
plt.title("Precision-Recall Curve")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df['metrics/mAP50(B)'])

plt.title("mAP50 Progress")
plt.xlabel("Epoch")
plt.ylabel("mAP50")

plt.show()
```

```python
model = YOLO("runs/detect/YOLOv11/weights/best.pt")

metrics = model.val(data=data_yaml)

conf = metrics.confusion_matrix.matrix

plt.figure(figsize=(8,6))

sns.heatmap(conf,annot=True,cmap="Blues")

plt.title("Confusion Matrix")
plt.show()
```

```python
results = model.predict(
    source=dataset_path+"/valid/images",
    save=True,
    conf=0.25
)
```
